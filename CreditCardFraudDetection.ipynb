# Credit Card Fraud Detection

## Problem Statement: 
    A Lot of credit card service providing companies in the current world are trying to improve their services for their customers, one of which is making sure that any fraudulent transaction is identified properly and resloved on time so that customers can have a hassle free experience and not get charged for any transaction they did not do.

## DataSet
    DataSet used here consists of 28 different dependent variables, a column consisting of the "Amount" for which the transaction has been done, and the last column "Class" categorizing the transaction into fraudulent or non-fraudulent. The very first column that can be ignored as it has nothing to with the categorization process, is the "Time" column, which we will be excluding while training and testing our selected classification model.

#### Import the Pycaret Classification Library

```python
from pycaret.classification import *
```

#### We remove the Time column as it has no effect in classifying a transaction as fraudulent or non-fraudulent, and divide the data into sampling data and testing data.

```python
df.drop(['Time'], axis=1, inplace=True)
data = df.sample(frac=0.95, random_state=786).reset_index(drop=True)
data_unseen = df.drop(data.index).reset_index(drop=True)

print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

#### Create a setup for pycaret to work on our data and find best algorithm

```python
clf = setup(data = data, target = 'Class', session_id=123) 
```

#### We use the compare_models() function to compare all the classification models to find best classifier model.

```python
compare_models()
```

#### We create the model that performs the best on our data.

```python
model = create_model('catboost')
```

#### We ensemble the model to check if it can perform any better by using model averaging method(background function of ensemble).

```python
CBEnsembleDef = ensemble_model(model)
```

#### Now we see all the graphs related to the performance of the ensmbled model.

```python
plot_model(CBEnsembleDef, 'confusion_matrix')
```

```python
plot_model(CBEnsembleDef, 'auc')
```

```python
plot_model(CBEnsembleDef, 'pr')
```

```python
plot_model(CBEnsembleDef, 'error')
```

```python
plot_model(CBEnsembleDef, 'class_report')
```

#### We optimize the threshold so as to reduce the risk of getting a false negetive as it will be the worst case scenario for our project.

```python
from pycaret.classification import *
```

```python
optimize_threshold(model, true_negative = 1500, false_negative = -5000)
```

#### Now we divide the data into train and test data again randomly so as to predict using the model and check how our model performs in random unknown conditions with input data.

```python
model_predict = predict_model(model)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df[df.columns[0:29]], df[df.columns[29]], test_size=0.2, random_state=0)
predictions = predict_model(model, x_test)
```
